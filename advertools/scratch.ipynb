```python

import re
import pandas as pd
def dataframify(func):
    def wrapper(*args):
        df = pd.DataFrame.from_records([func(*args)])
        return df
    
    return wrapper

@dataframify
def hashtag(regex, strings):
    return [re.findall(regex, s) for s in strings]

print(hashtag('#\w+', ['hello #there #how is this']))
```

```python
hashtag('@\w+', 'here is a @mention and a #hashtag')
```

```python
hashtags = [
    ['hello', 'hi'],
    ['hello', 'hey'],
    ['yo', 'what_is_this'],
    ['hello', 'great', 'good'],
]

from collections import defaultdict

dd = defaultdict(int)
lengths = defaultdict(int)

for hashes in hashtags:
    lengths[len(hashes)] += 1
    for hash in hashes:
        dd[hash] += 1

print(dd)
print(lengths)
```

```python
airline = pd.read_csv('/Users/Elias/Google Drive/socialmediadata-twitter-us-airline-sentiment/data/airline_sentiment_2_w_aa.csv',
                      encoding='utf-8')
airline.head(2)
```

```python
mentions = [re.findall('@\w+', x) for x in airline['text']]
```

```python
mentions_dd = defaultdict(int)
for mention in mentions:
    for m in mention:
        mentions_dd[m] += 1
print(*sorted(mentions_dd.items(), key=lambda x: x[1], reverse=True)[:10], sep='\n')
```

```python
hashtags = [re.findall('#\w+', x) for x in airline['text']]
hashtags_dd = defaultdict(int)
for hashtag in hashtags:
    for h in hashtag:
        hashtags_dd[h] += 1
print(*sorted(hashtags_dd.items(), key=lambda x: x[1], reverse=True)[:10], sep='\n')
```
